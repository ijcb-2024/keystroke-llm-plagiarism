```python
import os

folder_path = "/Users/debnathkundu/Downloads/Thesis/Dataset_Preparation/Buffalo/UB_keystroke_dataset"  # Replace with the actual folder path
unsegregated_=folder_path+"/Unsegregated"

file_list = []

for root, dirs, files in os.walk(unsegregated_):
    for file_name in files:
        file_path = os.path.join(root, file_name)
        file_list.append(file_path)

print(len(file_list))
file_list[0]
```

# The following cell segregates the files depending upon the keyboard used i.e. preparing them for keyboard specific/agnostic scenarios.
# Please note that the keyboard separated data is still in .txt format, it is converted to .json format through a different notebook named "cleaning_keyboard_txt_to_json.ipynb"

```python
import os
import shutil

# Replace with the actual folder path
new_folder_path = folder_path+"/Segregated"

# Create directories if they don't exist
os.makedirs(new_folder_path+"/K0/Free", exist_ok=True)
os.makedirs(new_folder_path+"/K0/Fixed", exist_ok=True)

os.makedirs(new_folder_path+"/K1/Free", exist_ok=True)
os.makedirs(new_folder_path+"/K1/Fixed", exist_ok=True)

os.makedirs(new_folder_path+"/K2/Free", exist_ok=True)
os.makedirs(new_folder_path+"/K2/Fixed", exist_ok=True)

os.makedirs(new_folder_path+"/K3/Free", exist_ok=True)
os.makedirs(new_folder_path+"/K3/Fixed", exist_ok=True)

for file_path in file_list:
    file_name = os.path.basename(file_path)
    if file_name[-6]=='0' and file_name[-5]=='0':
        shutil.copy(file_path, new_folder_path+"/K0/Fixed")
    elif file_name[-6]=='0' and file_name[-5]=='1':
        shutil.copy(file_path, new_folder_path+"/K0/Free")
        
    elif file_name[-6]=='1' and file_name[-5]=='0':
        shutil.copy(file_path, new_folder_path+"/K1/Fixed")
    elif file_name[-6]=='1' and file_name[-5]=='1':
        shutil.copy(file_path, new_folder_path+"/K1/Free")
        
    elif file_name[-6]=='2' and file_name[-5]=='0':
        shutil.copy(file_path, new_folder_path+"/K2/Fixed")
    elif file_name[-6]=='2' and file_name[-5]=='1':
        shutil.copy(file_path, new_folder_path+"/K2/Free")
        
    elif file_name[-6]=='3' and file_name[-5]=='0':
        shutil.copy(file_path, new_folder_path+"/K3/Fixed")
    elif file_name[-6]=='3' and file_name[-5]=='1':
        shutil.copy(file_path, new_folder_path+"/K3/Free")
        
```
